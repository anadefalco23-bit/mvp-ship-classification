## 1.3 Tipo de Problema

**Tipo escolhido:** Classificação Multiclasse Supervisionada.

**Justificativa:**

O problema investigado neste trabalho consiste na identificação automática da categoria de uma embarcação a partir de sua imagem, sendo esperado que cada amostra seja atribuída exclusivamente a uma única classe dentre as categorias previamente definidas no conjunto de dados. Dessa forma, o objetivo do modelo é aprender uma função de mapeamento entre imagens rotuladas e suas respectivas classes, caracterizando uma tarefa de aprendizado supervisionado.

Como o conjunto de dados é composto por múltiplas categorias mutuamente exclusivas de embarcações, o problema enquadra-se especificamente como uma tarefa de classificação multiclasse. Nessa abordagem, o modelo estima a probabilidade de pertencimento de cada imagem a todas as classes disponíveis, selecionando como resposta final aquela com maior probabilidade.

Esse tipo de problema é amplamente encontrado em aplicações de Visão Computacional, nas quais redes neurais profundas são treinadas para aprender automaticamente representações hierárquicas das imagens, dispensando a extração manual de atributos. No presente estudo, essa capacidade é explorada por meio da comparação entre diferentes arquiteturas de Deep Learning, incluindo uma CNN convencional, modelos baseados em Transfer Learning e a arquitetura YOLO11n-cls utilizada em modo de classificação.

A escolha pela classificação supervisionada também possibilita a utilização de métricas consolidadas para avaliação de desempenho, como acurácia, precisão, revocação e F1-score, permitindo comparar objetivamente a capacidade de generalização das diferentes arquiteturas analisadas. Dessa forma, torna-se possível identificar a solução mais adequada para aplicações de monitoramento marítimo baseadas em reconhecimento automático de padrões visuais.

## 2.2 Funções auxiliares

Com o objetivo de tornar o notebook mais organizado, modular e de fácil manutenção, foram desenvolvidas funções auxiliares para automatizar tarefas recorrentes ao longo do fluxo de trabalho. Essa estratégia reduz a repetição de código, melhora a legibilidade do projeto e facilita a reutilização dos procedimentos empregados nas diferentes etapas dos experimentos.

Além de contribuir para a organização do código, a utilização dessas funções garante que todos os modelos sejam avaliados utilizando exatamente os mesmos procedimentos e métricas, assegurando consistência metodológica e permitindo comparações justas entre a CNN Baseline, os modelos baseados em Transfer Learning (MobileNetV2 e ResNet50) e a arquitetura YOLO11n-cls.

```python
# ==========================================================
# Funções auxiliares
# ==========================================================

def show_images(images, labels, class_names, n=9):
    """
    Exibe exemplos de imagens do dataset.
    """

    plt.figure(figsize=(10, 10))

    for i in range(min(n, len(images))):
        plt.subplot(3, 3, i + 1)
        plt.imshow(images[i])

        # Caso os rótulos estejam em one-hot encoding
        if len(labels[i].shape) > 0:
            label_idx = np.argmax(labels[i])
        else:
            label_idx = int(labels[i])

        plt.title(class_names[label_idx])
        plt.axis("off")

    plt.tight_layout()
    plt.show()


def plot_training_history(history, model_name="Modelo"):
    """
    Exibe curvas de perda (loss) e acurácia para treino e validação
    de modelos treinados com TensorFlow/Keras.
    """

    fig, ax = plt.subplots(1, 2, figsize=(12, 4))

    ax[0].plot(history.history["loss"], label="Treino")
    ax[0].plot(history.history["val_loss"], label="Validação")
    ax[0].set_title(f"Loss - {model_name}")
    ax[0].set_xlabel("Época")
    ax[0].set_ylabel("Loss")
    ax[0].legend()

    ax[1].plot(history.history["accuracy"], label="Treino")
    ax[1].plot(history.history["val_accuracy"], label="Validação")
    ax[1].set_title(f"Accuracy - {model_name}")
    ax[1].set_xlabel("Época")
    ax[1].set_ylabel("Accuracy")
    ax[1].legend()

    plt.tight_layout()
    plt.show()


def get_predictions_from_keras_model(model, dataset):
    """
    Gera y_true e y_pred a partir de um modelo TensorFlow/Keras
    e de um tf.data.Dataset.
    """

    y_true = []
    y_pred = []

    for images, labels in dataset:
        predictions = model.predict(images, verbose=0)

        y_true.extend(np.argmax(labels.numpy(), axis=1))
        y_pred.extend(np.argmax(predictions, axis=1))

    return np.array(y_true), np.array(y_pred)


def evaluate_classification(y_true, y_pred, class_names=None):
    """
    Exibe métricas de classificação e retorna accuracy e F1-score macro.
    """

    print("Relatório de classificação:")
    print(
        classification_report(
            y_true,
            y_pred,
            target_names=class_names
        )
    )

    acc = accuracy_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred, average="macro")

    print(f"Accuracy: {acc:.4f}")
    print(f"F1-score Macro: {f1:.4f}")

    return acc, f1


def plot_confusion(y_true, y_pred, class_names):
    """
    Exibe a matriz de confusão.
    """

    cm = confusion_matrix(y_true, y_pred)

    plt.figure(figsize=(8, 6))

    sns.heatmap(
        cm,
        annot=True,
        fmt="d",
        cmap="Blues",
        xticklabels=class_names,
        yticklabels=class_names
    )

    plt.xlabel("Previsto")
    plt.ylabel("Real")
    plt.title("Matriz de Confusão")
    plt.tight_layout()
    plt.show()
```

# 5. Preparação dos Dados e Divisão dos Conjuntos

Nesta etapa são definidos os dados de entrada utilizados pelos modelos, a variável-alvo do problema e a estratégia de divisão do conjunto de dados para treinamento, validação e teste.

Diferentemente de problemas tradicionais de Machine Learning baseados em tabelas, nos quais as variáveis preditoras são explicitamente selecionadas pelo pesquisador, neste trabalho as características discriminativas são aprendidas automaticamente a partir dos próprios pixels das imagens por arquiteturas de Deep Learning. Dessa forma, não há necessidade de engenharia manual de atributos, uma vez que os modelos extraem representações hierárquicas durante o processo de treinamento.

As imagens digitais constituem as entradas (*inputs*) dos modelos, enquanto a coluna **category** representa a variável-alvo (*target*), indicando a categoria da embarcação associada a cada imagem. As variáveis auxiliares criadas durante a análise exploratória — como **image_path**, **file_exists**, **width**, **height** e **mode** — foram empregadas exclusivamente para validação e caracterização do conjunto de dados, não sendo utilizadas diretamente durante o treinamento das redes neurais.

O carregamento das imagens é realizado por meio dos mecanismos de processamento de dados disponibilizados pelas bibliotecas TensorFlow/Keras e Ultralytics, permitindo leitura eficiente dos arquivos, redimensionamento automático e preparação dos lotes (*batches*) utilizados durante o treinamento. Essa abordagem reduz o consumo de memória e garante maior reprodutibilidade dos experimentos.

Para avaliação adequada do desempenho dos modelos, o conjunto de dados foi dividido em subconjuntos independentes de treinamento, validação e teste. O conjunto de treinamento é utilizado para o ajuste dos parâmetros das arquiteturas avaliadas, o conjunto de validação é empregado para monitorar o processo de aprendizado, comparar modelos e selecionar hiperparâmetros, enquanto o conjunto de teste permanece completamente isolado durante essas etapas, sendo utilizado exclusivamente para a avaliação final da capacidade de generalização.

Como o problema não apresenta dependência temporal entre as observações, adotou-se uma estratégia de divisão aleatória estratificada. A estratificação preserva aproximadamente a mesma proporção de imagens de cada categoria nos diferentes subconjuntos, reduzindo possíveis vieses decorrentes do leve desbalanceamento identificado durante a análise exploratória e permitindo comparações mais confiáveis entre as arquiteturas CNN Baseline, MobileNetV2, ResNet50 e YOLO11n-cls.

## 5.1 Justificativa da Divisão

A estratégia de divisão dos dados adotada neste trabalho baseia-se no método **holdout estratificado**, amplamente empregado em problemas de classificação de imagens utilizando Deep Learning. Essa abordagem consiste em separar o conjunto de dados em subconjuntos independentes de treinamento, validação e teste, permitindo que cada etapa do desenvolvimento dos modelos seja conduzida sobre dados distintos e sem sobreposição de informações.

O conjunto de treinamento foi utilizado para o ajuste dos parâmetros das arquiteturas avaliadas, enquanto o conjunto de validação desempenhou papel fundamental no monitoramento do processo de aprendizado, na comparação entre os modelos candidatos e na seleção de hiperparâmetros. O conjunto de teste permaneceu completamente isolado durante essas etapas, sendo utilizado exclusivamente para a avaliação final da capacidade de generalização do modelo selecionado.

Optou-se por uma divisão de **70% para treinamento, 15% para validação e 15% para teste**, proporção amplamente adotada em aplicações de Visão Computacional. Essa configuração oferece um equilíbrio adequado entre disponibilidade de dados para aprendizado e robustez estatística da avaliação, preservando quantidade suficiente de imagens para o treinamento sem comprometer a representatividade dos conjuntos de validação e teste.

A divisão foi realizada de forma **estratificada** em relação à variável-alvo (**category**), garantindo que todas as categorias de embarcações permanecessem representadas em proporções semelhantes nos diferentes subconjuntos. Essa estratégia minimiza possíveis vieses decorrentes do leve desbalanceamento observado durante a análise exploratória e assegura que todas as arquiteturas avaliadas sejam treinadas e comparadas sob condições equivalentes.

Outro aspecto importante da estratégia adotada é a prevenção do **vazamento de dados (*data leakage*)**. Como nenhuma imagem pertencente aos conjuntos de validação ou teste participa do treinamento dos modelos, as métricas obtidas refletem de maneira mais confiável a capacidade de generalização das arquiteturas avaliadas para novas imagens de embarcações. Esse cuidado metodológico aumenta a confiabilidade dos experimentos e fortalece a validade das comparações realizadas entre a CNN Baseline, MobileNetV2, ResNet50 e YOLO11n-cls.


# 6. Pré-processamento e Pipeline

Nesta etapa, as imagens foram preparadas para utilização pelas arquiteturas de Deep Learning avaliadas neste trabalho. Diferentemente de problemas tradicionais de Machine Learning baseados em dados tabulares, o pré-processamento concentrou-se na padronização das imagens e na construção de um fluxo de processamento compatível com modelos de Visão Computacional.

Inicialmente, todas as imagens foram carregadas diretamente a partir de seus respectivos arquivos e convertidas para o espaço de cores RGB, garantindo compatibilidade com as arquiteturas pré-treinadas utilizadas neste estudo. Em seguida, todas as imagens foram redimensionadas para uma resolução fixa de 224 × 224 pixels, assegurando que todas as entradas apresentassem dimensões uniformes durante o treinamento.

Posteriormente, os valores dos pixels foram normalizados para a faixa entre 0 e 1, procedimento que contribui para maior estabilidade numérica durante o processo de otimização e favorece a convergência das redes neurais.

A variável-alvo (category) foi inicialmente ajustada para iniciar em zero (classes 0 a 4) e, posteriormente, convertida para representação categórica (one-hot encoding), tornando-a compatível com a função de perda Categorical Crossentropy empregada durante o treinamento dos modelos implementados em TensorFlow/Keras.

As imagens foram organizadas em lotes (batches), permitindo processamento eficiente durante o treinamento em GPU e reduzindo o consumo de memória. Além disso, os conjuntos de treinamento, validação e teste permaneceram completamente independentes durante todo o fluxo de processamento, evitando qualquer forma de vazamento de informações (data leakage).

Como etapa complementar, foi realizada uma estratégia de Fine Tuning sobre a arquitetura ResNet50. Nessa etapa, parte das camadas convolucionais previamente congeladas foi liberada para treinamento utilizando uma taxa de aprendizado reduzida, permitindo adaptar os pesos pré-treinados às características específicas do conjunto de imagens de embarcações. Essa abordagem buscou aumentar a capacidade de generalização do modelo sem comprometer o conhecimento previamente adquirido durante o treinamento no ImageNet.

O pipeline descrito foi empregado na implementação dos modelos desenvolvidos em TensorFlow/Keras (CNN Baseline, MobileNetV2 e ResNet50). Para o YOLO11n-cls, o pré-processamento equivalente foi executado automaticamente pela biblioteca Ultralytics, preservando a resolução de entrada e garantindo consistência entre os diferentes experimentos realizados.

| Etapa                                   | Objetivo                                                                |
| --------------------------------------- | ----------------------------------------------------------------------- |
| Verificação da integridade dos arquivos | Garantir que todas as imagens estejam disponíveis para uso              |
| Conversão para RGB                      | Padronizar o número de canais e garantir compatibilidade com os modelos |
| Redimensionamento (224 × 224)           | Uniformizar as dimensões das imagens de entrada                         |
| Normalização dos pixels (0–1)           | Melhorar a estabilidade numérica e favorecer a convergência             |

```python
# ==========================================================
# 6. Pré-processamento e Pipeline
# ==========================================================
# Definição dos principais parâmetros utilizados no
# processamento das imagens e no treinamento dos modelos.
# ==========================================================

# Tamanho padrão das imagens.
# O formato 224x224 é compatível com MobileNetV2, ResNet50
# e também foi utilizado no treinamento do YOLO11n-cls.
IMG_SIZE = (224, 224)

# Número de imagens processadas por lote durante o treinamento.
# Um batch size de 32 oferece bom equilíbrio entre desempenho
# e uso de memória no ambiente Google Colab.
BATCH_SIZE = 32

# Número de categorias do problema de classificação.
NUM_CLASSES = df[TARGET].nunique()

# Lista ordenada dos códigos das categorias
CLASS_IDS = sorted(df[TARGET].unique())

# Lista com os nomes das categorias
CLASS_LABELS = [class_names[i] for i in CLASS_IDS]

print("Tamanho das imagens:", IMG_SIZE)
print("Batch size:", BATCH_SIZE)
print("Número de classes:", NUM_CLASSES)
print("Categorias:", CLASS_LABELS)
```

```python
# ==========================================================
# Data Augmentation
# ==========================================================
# Durante o treinamento, algumas transformações aleatórias
# são aplicadas às imagens com o objetivo de aumentar a
# variabilidade dos dados e reduzir o risco de overfitting.
#
# Essas transformações são aplicadas apenas ao conjunto de
# treinamento e não modificam as imagens de validação ou teste.
# ==========================================================

from tensorflow.keras import Sequential
from tensorflow.keras.layers import (
    RandomFlip,
    RandomRotation,
    RandomZoom
)

data_augmentation = Sequential(

    [

        # Espelhamento horizontal aleatório
        RandomFlip("horizontal"),

        # Rotação aleatória de até aproximadamente ±10%
        RandomRotation(0.10),

        # Zoom aleatório de até 10%
        RandomZoom(0.10)

    ],

    name="Data_Augmentation"

)

print("Pipeline de Data Augmentation criado com sucesso.")
```

```python
# ==========================================================
# Conversão dos rótulos para índices numéricos
# ==========================================================
# O TensorFlow/Keras utiliza, neste projeto, codificação das
# classes iniciando em zero (0).
#
# Como o dataset original identifica as categorias de 1 a 5,
# cria-se uma nova coluna denominada "label", contendo os
# índices de 0 a 4. A coluna original ("category") é mantida
# para facilitar a interpretação dos resultados e a análise
# exploratória.
# ==========================================================

# Cria cópias independentes dos conjuntos de dados
# para evitar alterações acidentais no DataFrame original.
train_df = train_df.copy()
val_df   = val_df.copy()
test_df  = test_df.copy()

# Converte as categorias de 1–5 para 0–4
for dataset in [train_df, val_df, test_df]:
    dataset["label"] = dataset["category"] - 1

# Verifica se todas as classes foram convertidas corretamente
print("Classes no conjunto de treino:", sorted(train_df["label"].unique()))
print("Classes no conjunto de validação:", sorted(val_df["label"].unique()))
print("Classes no conjunto de teste:", sorted(test_df["label"].unique()))
```

```python
# ==========================================================
# Correspondência entre rótulos originais e índices internos
# ==========================================================

label_mapping = {
    categoria - 1: nome
    for categoria, nome in class_names.items()
}

print("Mapeamento das classes utilizado pelos modelos:")

for indice, nome in label_mapping.items():
    print(f"{indice} -> {nome}")
```

```python
# ==========================================================
# Função para carregamento e pré-processamento das imagens
# ==========================================================
# Esta função implementa o pipeline de processamento utilizado
# pelos modelos desenvolvidos em TensorFlow/Keras
# (CNN Baseline, MobileNetV2 e ResNet50).
#
# Para cada imagem são executadas as seguintes etapas:
# 1. leitura do arquivo;
# 2. conversão para RGB;
# 3. redimensionamento para 224x224 pixels;
# 4. normalização dos pixels;
# 5. conversão do rótulo para One-Hot Encoding.
# ==========================================================

def load_and_preprocess_image(image_path, label):

    """
    Carrega uma imagem e aplica todas as etapas de
    pré-processamento necessárias para o treinamento
    dos modelos de Deep Learning.

    Parâmetros
    ----------
    image_path : str
        Caminho completo da imagem.

    label : int
        Classe da embarcação (0 a NUM_CLASSES-1).

    Retorno
    -------
    image : Tensor
        Imagem pré-processada.

    label : Tensor
        Rótulo convertido para One-Hot Encoding.
    """

    # ======================================================
    # Leitura da imagem
    # ======================================================

    image = tf.io.read_file(image_path)

    # ======================================================
    # Decodificação
    # ======================================================
    # Todas as imagens são convertidas para RGB
    # (3 canais), garantindo compatibilidade com
    # as arquiteturas pré-treinadas.

    image = tf.image.decode_image(
        image,
        channels=3,
        expand_animations=False
    )

    # ======================================================
    # Redimensionamento
    # ======================================================
    # Todas as imagens passam a possuir resolução
    # padronizada de 224x224 pixels.

    image = tf.image.resize(
        image,
        IMG_SIZE
    )

    # ======================================================
    # Conversão e normalização
    # ======================================================
    # Converte automaticamente para float32 e
    # normaliza os pixels para o intervalo [0,1].

    image = tf.image.convert_image_dtype(
        image,
        tf.float32
    )

    # ======================================================
    # Conversão do rótulo
    # ======================================================
    # O TensorFlow trabalha com classificação
    # multiclasse utilizando One-Hot Encoding.

    label = tf.one_hot(
        label,
        depth=NUM_CLASSES
    )

    return image, label

    # ==========================================================
# Observação
# ==========================================================
# Este pipeline é utilizado exclusivamente pelos modelos
# implementados em TensorFlow/Keras (CNN Baseline,
# MobileNetV2 e ResNet50).
#
# O modelo YOLO11n-cls possui seu próprio pipeline interno
# de carregamento e pré-processamento de imagens, conforme
# implementado pela biblioteca Ultralytics.
# ==========================================================
```

```python
# ==========================================================
# Função para criação de datasets TensorFlow
# ==========================================================
# Esta função cria objetos tf.data.Dataset a partir dos
# DataFrames de treino, validação e teste.
#
# O pipeline aplica:
# - leitura dos caminhos das imagens;
# - associação com os rótulos;
# - pré-processamento;
# - embaralhamento opcional;
# - organização em batches;
# - prefetch para otimização do desempenho.
# ==========================================================

def create_tf_dataset(dataframe, shuffle=False):
    """
    Cria um objeto tf.data.Dataset a partir de um DataFrame
    contendo caminhos das imagens e rótulos numéricos.

    Parâmetros
    ----------
    dataframe : pandas.DataFrame
        DataFrame contendo as colunas image_path e label.

    shuffle : bool
        Indica se o dataset deve ser embaralhado.
        Deve ser True apenas para o conjunto de treinamento.

    Retorno
    -------
    dataset : tf.data.Dataset
        Dataset TensorFlow pronto para treinamento ou avaliação.
    """

    # Caminhos das imagens
    image_paths = dataframe[IMAGE_PATH_COL].values

    # Rótulos numéricos já ajustados para iniciar em zero
    labels = dataframe["label"].values

    # Cria o dataset TensorFlow
    dataset = tf.data.Dataset.from_tensor_slices(
        (image_paths, labels)
    )

    # Aplica o pré-processamento às imagens
    dataset = dataset.map(
        load_and_preprocess_image,
        num_parallel_calls=tf.data.AUTOTUNE
    )

    # Embaralha apenas o conjunto de treinamento
    if shuffle:
        dataset = dataset.shuffle(
            buffer_size=len(dataframe),
            seed=SEED,
            reshuffle_each_iteration=True
        )

    # Agrupa os dados em batches
    dataset = dataset.batch(BATCH_SIZE)

    # Otimiza o carregamento durante o treinamento
    dataset = dataset.prefetch(tf.data.AUTOTUNE)

    return dataset
```

```python
# ==========================================================
# Criação dos datasets TensorFlow
# ==========================================================
# Nesta etapa são criados os objetos tf.data.Dataset que
# serão utilizados durante o treinamento, validação e teste
# dos modelos implementados em TensorFlow/Keras.
#
# Apenas o conjunto de treinamento é embaralhado, garantindo
# maior aleatoriedade durante o aprendizado sem comprometer
# a reprodutibilidade dos experimentos.
# ==========================================================

# Conjunto de treinamento
train_ds = create_tf_dataset(
    train_df,
    shuffle=True
)

# Conjunto de validação
val_ds = create_tf_dataset(
    val_df,
    shuffle=False
)

# Conjunto de teste
test_ds = create_tf_dataset(
    test_df,
    shuffle=False
)

# ==========================================================
# Resumo dos datasets criados
# ==========================================================

print("Datasets TensorFlow criados com sucesso!\n")

print(f"Treino    : {len(train_df):,} imagens")
print(f"Validação : {len(val_df):,} imagens")
print(f"Teste     : {len(test_df):,} imagens")

print(f"\nBatch size: {BATCH_SIZE}")
print(f"Número de classes: {NUM_CLASSES}")

# ==========================================================
# Observação
# ==========================================================
# Os datasets criados nesta etapa são utilizados
# exclusivamente pelos modelos desenvolvidos em
# TensorFlow/Keras (CNN Baseline, MobileNetV2 e ResNet50).
#
# O modelo YOLO11n-cls utiliza um pipeline próprio de
# carregamento e processamento de imagens implementado
# pela biblioteca Ultralytics.
# ==========================================================
```

Nesse contexto, foi adotada uma estratégia comparativa composta por um modelo baseline e três arquiteturas candidatas amplamente utilizadas em aplicações de Visão Computacional, permitindo investigar diferentes abordagens para a classificação automática de embarcações.

O modelo baseline consiste em uma Rede Neural Convolucional (CNN) desenvolvida especificamente para este trabalho. Sua principal finalidade é estabelecer uma referência de desempenho (baseline), permitindo quantificar os ganhos obtidos por arquiteturas mais sofisticadas baseadas em Transfer Learning e modelos modernos de classificação.

Sua utilização permite investigar o impacto do reaproveitamento de representações visuais previamente aprendidas em um modelo compacto e eficiente, frequentemente utilizado em dispositivos embarcados e aplicações em tempo real.

O segundo modelo candidato corresponde à ResNet50, também pré-treinada no ImageNet. Como terceira arquitetura candidata foi incorporado o YOLO11n-cls, disponibilizado pela biblioteca Ultralytics. Embora a família YOLO tenha sido originalmente desenvolvida para detecção de objetos, a versão YOLO11n-cls foi projetada especificamente para tarefas de classificação de imagens, incorporando avanços recentes em arquitetura de redes neurais convolucionais e estratégias modernas de treinamento. Sua inclusão amplia o escopo do estudo ao permitir a comparação entre arquiteturas clássicas implementadas em TensorFlow/Keras e uma arquitetura contemporânea otimizada para classificação visual.

As quatro arquiteturas avaliadas representam diferentes níveis de complexidade e distintas estratégias de aprendizado. A CNN Baseline permite avaliar o desempenho de uma rede construída especificamente para o problema; a MobileNetV2 representa uma abordagem eficiente baseada em Transfer Learning; a ResNet50 investiga os benefícios de arquiteturas profundas; e o YOLO11n-cls incorpora uma solução moderna voltada para classificação de imagens. Essa diversidade de modelos possibilita uma comparação abrangente entre desempenho preditivo, capacidade de generalização, custo computacional e potencial de aplicação em sistemas de monitoramento marítimo.




```python
# ==========================================================
# Modelo Baseline
# Rede Neural Convolucional (CNN)
# ==========================================================
# Esta arquitetura foi desenvolvida especificamente para este
# projeto e constitui o modelo de referência (baseline).
#
# Seu objetivo é estabelecer uma base de comparação para as
# arquiteturas mais sofisticadas avaliadas posteriormente,
# como MobileNetV2, ResNet50 e YOLO11n-cls.
# ==========================================================

from tensorflow.keras import Sequential
from tensorflow.keras.layers import (
    Conv2D,
    MaxPooling2D,
    Flatten,
    Dense,
    Dropout
)

baseline_model = Sequential([

    # ======================================================
    # (Opcional) Data Augmentation
    # ======================================================
    # O Data Augmentation foi avaliado em experimentos
    # específicos. Entretanto, o modelo baseline utilizado
    # para comparação foi treinado sem aumento artificial
    # das imagens.
    #
    # data_augmentation,

    # ======================================================
    # Primeiro bloco convolucional
    # ======================================================

    Conv2D(
        filters=32,
        kernel_size=(3,3),
        activation="relu",
        input_shape=(*IMG_SIZE, 3)
    ),

    MaxPooling2D(),

    # ======================================================
    # Segundo bloco convolucional
    # ======================================================

    Conv2D(
        filters=64,
        kernel_size=(3,3),
        activation="relu"
    ),

    MaxPooling2D(),

    # ======================================================
    # Terceiro bloco convolucional
    # ======================================================

    Conv2D(
        filters=128,
        kernel_size=(3,3),
        activation="relu"
    ),

    MaxPooling2D(),

    # ======================================================
    # Camadas totalmente conectadas
    # ======================================================

    Flatten(),

    Dense(
        128,
        activation="relu"
    ),

    # Dropout para reduzir overfitting
    Dropout(0.30),

    # Camada de saída
    Dense(
        NUM_CLASSES,
        activation="softmax"
    )

])

print("Modelo Baseline criado com sucesso.")
```

A arquitetura proposta possui aproximadamente 11,17 milhões de parâmetros treináveis, concentrados principalmente na camada totalmente conectada após a operação de Flatten. Embora apresente maior custo computacional em relação às arquiteturas pré-treinadas, o modelo fornece uma referência importante para avaliar os ganhos obtidos pelas abordagens mais sofisticadas.

```python
# ==========================================================
# Modelo Candidato 1: MobileNetV2
# Transfer Learning sem Data Augmentation
# ==========================================================
# A MobileNetV2 é utilizada como extratora de características,
# com pesos pré-treinados no ImageNet.
#
# Nesta configuração inicial, as camadas convolucionais são
# mantidas congeladas, e apenas a camada classificadora final
# é treinada para o problema de classificação de embarcações.
# ==========================================================

from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense
from tensorflow.keras.models import Model

# Carrega a MobileNetV2 pré-treinada no ImageNet,
# removendo a camada classificadora original.
base_mobilenet = MobileNetV2(
    weights="imagenet",
    include_top=False,
    input_shape=(*IMG_SIZE, 3)
)

# Congela o backbone convolucional
base_mobilenet.trainable = False

# Entrada do modelo
inputs = base_mobilenet.input

# Extração de características
x = base_mobilenet(
    inputs,
    training=False
)

# Agregação dos mapas de características
x = GlobalAveragePooling2D()(x)

# Camada final de classificação
outputs = Dense(
    NUM_CLASSES,
    activation="softmax"
)(x)

# Modelo final
mobilenet_model = Model(
    inputs=inputs,
    outputs=outputs,
    name="MobileNetV2_Transfer_Learning"
)

print("Modelo MobileNetV2 criado com sucesso.")

mobilenet_model.summary()
```

```python
# ==========================================================
# Modelo Candidato 2: ResNet50
# Transfer Learning sem Data Augmentation
# ==========================================================
# A ResNet50 é utilizada como extratora de características,
# com pesos pré-treinados no ImageNet.
#
# Nesta configuração inicial, a base convolucional permanece
# congelada, sendo treinada apenas a camada classificadora
# final adaptada ao problema de classificação de embarcações.
# ==========================================================

from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense
from tensorflow.keras.models import Model

# Carrega a ResNet50 pré-treinada no ImageNet,
# removendo a camada classificadora original.
base_resnet = ResNet50(
    weights="imagenet",
    include_top=False,
    input_shape=(*IMG_SIZE, 3)
)

# Congela o backbone convolucional
base_resnet.trainable = False

# Entrada do modelo
inputs = base_resnet.input

# Extração de características
x = base_resnet(
    inputs,
    training=False
)

# Agregação dos mapas de características
x = GlobalAveragePooling2D()(x)

# Camada final de classificação
outputs = Dense(
    NUM_CLASSES,
    activation="softmax"
)(x)

# Modelo final
resnet_model = Model(
    inputs=inputs,
    outputs=outputs,
    name="ResNet50_Transfer_Learning"
)

print("Modelo ResNet50 criado com sucesso.")

resnet_model.summary()
```

```python
# ==========================================================
# Configuração comum de treinamento
# ==========================================================
# Estas configurações serão utilizadas por todos os modelos
# implementados em TensorFlow/Keras, garantindo padronização
# e comparação justa entre as arquiteturas.
# ==========================================================

OPTIMIZER = "adam"
LOSS_FUNCTION = "categorical_crossentropy"
METRICS = ["accuracy"]


# ==========================================================
# Compilação dos modelos TensorFlow/Keras
# ==========================================================
# Apenas os modelos implementados em TensorFlow/Keras
# necessitam desta etapa.
#
# O modelo YOLO11n-cls utiliza a biblioteca Ultralytics,
# que possui pipeline próprio de treinamento e, portanto,
# não requer compilação explícita.
# ==========================================================

models = {
    "CNN Baseline": baseline_model,
    "MobileNetV2": mobilenet_model,
    "ResNet50": resnet_model
}

for name, model in models.items():

    model.compile(
        optimizer=OPTIMIZER,
        loss=LOSS_FUNCTION,
        metrics=METRICS
    )

    print(f"✓ {name} compilado com sucesso.")

print("\nTodos os modelos TensorFlow/Keras foram compilados com sucesso.")
```

# 8. Treinamento e Avaliação dos Modelos

Após a definição das arquiteturas candidatas, foi realizada a etapa de treinamento e avaliação dos modelos de classificação. Todos os experimentos foram conduzidos utilizando os mesmos subconjuntos de treinamento, validação e teste, preservando a divisão estratificada estabelecida anteriormente e garantindo comparabilidade entre os resultados obtidos.

Para os modelos implementados em TensorFlow/Keras (CNN Baseline, MobileNetV2 e ResNet50), foram adotadas configurações comuns de treinamento, incluindo o mesmo otimizador, função de perda, métrica principal e número máximo de épocas. Além disso, foram utilizados mecanismos de interrupção antecipada (*Early Stopping*) e redução adaptativa da taxa de aprendizado (*ReduceLROnPlateau*), com o objetivo de evitar sobreajuste e otimizar o processo de treinamento.

O modelo YOLO11n-cls foi treinado utilizando o pipeline disponibilizado pela biblioteca Ultralytics, que possui mecanismos próprios de otimização, monitoramento e salvamento do melhor modelo durante o treinamento. Embora empregue uma estratégia de treinamento distinta daquela utilizada pelos modelos TensorFlow/Keras, foram mantidos os mesmos conjuntos de treinamento, validação e teste, permitindo uma comparação objetiva entre as arquiteturas avaliadas.

Ao final desta etapa, todos os modelos serão comparados utilizando métricas quantitativas e análises qualitativas, permitindo identificar a arquitetura mais adequada para o problema de classificação automática de embarcações.


```python
# ==========================================================
# 8.1 Configuração comum de treinamento
# ==========================================================
# Estas configurações serão utilizadas pelos modelos
# implementados em TensorFlow/Keras:
# - CNN Baseline;
# - MobileNetV2;
# - ResNet50.
#
# O YOLO11n-cls utiliza pipeline próprio da biblioteca
# Ultralytics e será treinado em uma etapa separada.
# ==========================================================

import time

from tensorflow.keras.callbacks import (
    EarlyStopping,
    ReduceLROnPlateau
)

# Número máximo de épocas
EPOCHS = 10

# Early Stopping para interromper o treinamento caso
# não haja melhora na perda de validação.
early_stopping = EarlyStopping(
    monitor="val_loss",
    patience=3,
    restore_best_weights=True
)

# Redução automática da taxa de aprendizado quando
# a perda de validação deixa de melhorar.
reduce_lr = ReduceLROnPlateau(
    monitor="val_loss",
    factor=0.2,
    patience=2,
    verbose=1
)

# Lista de callbacks comum aos modelos TensorFlow/Keras
callbacks = [
    early_stopping,
    reduce_lr
]

print("Configuração comum de treinamento criada com sucesso.")
```

```python
# ==========================================================
# 8.2 Função auxiliar para treinamento dos modelos Keras
# ==========================================================
# Esta função centraliza o treinamento dos modelos
# TensorFlow/Keras, reduzindo repetição de código e
# garantindo que todas as arquiteturas sejam treinadas
# sob as mesmas condições experimentais.
# ==========================================================

def train_model(
    model,
    model_name,
    train_dataset,
    validation_dataset,
    epochs=EPOCHS,
    callbacks_list=callbacks
):
    """
    Treina um modelo TensorFlow/Keras e registra o tempo total
    de treinamento.

    Parâmetros
    ----------
    model : tf.keras.Model
        Modelo a ser treinado.

    model_name : str
        Nome do modelo utilizado nos registros.

    train_dataset : tf.data.Dataset
        Dataset de treinamento.

    validation_dataset : tf.data.Dataset
        Dataset de validação.

    epochs : int
        Número máximo de épocas.

    callbacks_list : list
        Lista de callbacks utilizados durante o treinamento.

    Retorno
    -------
    history : keras.callbacks.History
        Histórico de treinamento do modelo.

    training_time : float
        Tempo total de treinamento, em segundos.
    """

    print("=" * 60)
    print(f"Iniciando treinamento do modelo: {model_name}")
    print("=" * 60)

    start_time = time.time()

    history = model.fit(
        train_dataset,
        validation_data=validation_dataset,
        epochs=epochs,
        callbacks=callbacks_list,
        verbose=1
    )

    end_time = time.time()

    training_time = end_time - start_time

    print("\n" + "-" * 60)
    print(f"Treinamento finalizado: {model_name}")
    print(f"Tempo total: {training_time:.2f} segundos")
    print("-" * 60)

    return history, training_time
```

```python
# ==========================================================
# 8.3 Treinamento dos modelos TensorFlow/Keras
# ==========================================================
# Os três modelos são treinados utilizando a mesma função,
# os mesmos datasets, o mesmo número máximo de épocas e os
# mesmos callbacks.
# ==========================================================

# Dicionários para armazenar os históricos e tempos de treino
histories = {}
training_times = {}

for model_name, model in models.items():

    histories[model_name], training_times[model_name] = train_model(
        model=model,
        model_name=model_name,
        train_dataset=train_ds,
        validation_dataset=val_ds
    )

print("\nTreinamento dos modelos TensorFlow/Keras concluído.")
```

```python
def plot_training_history(history, model_name="Modelo"):
    """
    Exibe curvas de loss e accuracy para treino e validação.
    """

    fig, axes = plt.subplots(1, 2, figsize=(12, 4))

    # Loss
    axes[0].plot(history.history["loss"], label="Treino")
    axes[0].plot(history.history["val_loss"], label="Validação")
    axes[0].set_title(f"Loss - {model_name}")
    axes[0].set_xlabel("Época")
    axes[0].set_ylabel("Loss")
    axes[0].legend()
    axes[0].grid(True)

    # Accuracy
    axes[1].plot(history.history["accuracy"], label="Treino")
    axes[1].plot(history.history["val_accuracy"], label="Validação")
    axes[1].set_title(f"Accuracy - {model_name}")
    axes[1].set_xlabel("Época")
    axes[1].set_ylabel("Accuracy")
    axes[1].legend()
    axes[1].grid(True)

    plt.tight_layout()
    plt.show()
```

```python
# ==========================================================
# Curvas de treinamento e validação dos modelos TensorFlow/Keras
# ==========================================================
# Nesta etapa são exibidas as curvas de loss e accuracy dos
# modelos implementados em TensorFlow/Keras.
#
# O YOLO11n-cls utiliza o pipeline da Ultralytics e gera seus
# próprios gráficos de treinamento automaticamente no diretório:
# /content/yolo_results/YOLO11n_ship_classification
# ==========================================================

for name, history in histories.items():

    print("=" * 60)
    print(f"Curvas de treinamento do modelo: {name}")
    print("=" * 60)

    plot_training_history(
        history,
        model_name=name
    )
```

# 9. Validação e Otimização de Hiperparâmetros

Após a avaliação inicial dos modelos, a ResNet50 foi selecionada para a etapa de otimização por apresentar o melhor desempenho entre as arquiteturas implementadas em TensorFlow/Keras. Embora o modelo YOLO11n-cls tenha obtido a maior acurácia geral, seu treinamento é realizado por meio da biblioteca Ultralytics, que incorpora automaticamente diversas estratégias internas de otimização de hiperparâmetros. Dessa forma, optou-se por realizar a otimização apenas sobre a ResNet50, permitindo uma comparação mais consistente entre os modelos desenvolvidos no ambiente TensorFlow/Keras.

A estratégia adotada consistiu na aplicação de Fine Tuning, técnica amplamente empregada em Transfer Learning. Inicialmente, todas as camadas convolucionais permaneceram congeladas durante o treinamento da camada classificadora. Em seguida, parte das camadas finais da ResNet50 foi descongelada, permitindo um ajuste fino dos pesos previamente aprendidos no ImageNet às características específicas das imagens de embarcações.

Durante essa etapa foi utilizada uma taxa de aprendizado reduzida, de forma a preservar o conhecimento previamente adquirido pela rede e evitar alterações excessivas nos pesos já treinados. Essa estratégia favorece uma adaptação gradual ao novo domínio de aplicação, reduzindo riscos de sobreajuste e instabilidade durante o treinamento.

A validação continuou sendo realizada utilizando exclusivamente o conjunto de validação previamente separado, mantendo o conjunto de teste completamente isolado para a avaliação final dos modelos. Essa separação evita vazamento de informações (data leakage) e fornece uma estimativa mais confiável da capacidade de generalização.

Como métrica principal de comparação foi utilizada a acurácia de validação, complementada posteriormente por métricas como precisão (precision), revocação (recall), F1-Score e matriz de confusão, proporcionando uma análise mais abrangente do desempenho obtido.

## 9.1 Discussão da Otimização

Embora a otimização de hiperparâmetros constitua uma etapa importante no desenvolvimento de modelos de Deep Learning, sua aplicação deve ser compatível com os objetivos e o escopo do projeto. Em vez de concentrar esforços na busca exaustiva por ganhos marginais de desempenho, priorizou-se a comparação sistemática entre arquiteturas representativas, permitindo identificar diferenças significativas entre uma CNN treinada do zero, modelos clássicos de Transfer Learning e uma arquitetura moderna baseada na família YOLO.

```python
# ==========================================================
# 10.2 Exemplos de classificações incorretas
# ==========================================================
# Nesta etapa são apresentadas algumas imagens do conjunto
# de teste classificadas incorretamente pelo modelo YOLO11n-cls.
#
# A análise qualitativa desses exemplos auxilia na
# compreensão dos principais padrões de erro observados.
# ==========================================================

from pathlib import Path
from PIL import Image
import matplotlib.pyplot as plt

errors = []

# Percorre todas as classes do conjunto de teste
for true_idx, class_name in enumerate(yolo_class_order):

    class_dir = YOLO_DATASET_DIR / "test" / class_name

    image_files = list(class_dir.glob("*"))

    for image_path in image_files:

        result = best_yolo.predict(
            source=str(image_path),
            imgsz=224,
            verbose=False
        )[0]

        pred_idx = int(result.probs.top1)

        if pred_idx != true_idx:

            confidence = float(result.probs.top1conf)

            errors.append({

                "path": image_path,

                "real": class_name,

                "pred": yolo_class_order[pred_idx],

                "confidence": confidence

            })

print("="*60)
print(f"Total de classificações incorretas: {len(errors)}")
print("="*60)
```

```python
# ==========================================================
# Visualização dos erros
# ==========================================================

n = min(9, len(errors))

plt.figure(figsize=(13,13))

for i in range(n):

    error = errors[i]

    img = Image.open(error["path"])

    plt.subplot(3,3,i+1)

    plt.imshow(img)

    plt.title(

        f"Real: {error['real']}\n"

        f"Prev.: {error['pred']}\n"

        f"Conf.: {error['confidence']:.2%}",

        fontsize=10

    )

    plt.axis("off")

plt.suptitle(

    "Exemplos de Classificações Incorretas - YOLO11n-cls",

    fontsize=15,

    fontweight="bold"

)

plt.tight_layout()

plt.show()
```

# 15. Apêndice — Resumo Técnico das Arquiteturas Avaliadas

Este apêndice apresenta uma visão resumida das arquiteturas de Deep Learning investigadas ao longo do projeto, destacando suas principais características, vantagens, limitações e o desempenho observado durante os experimentos. Seu objetivo é complementar a discussão apresentada nas seções anteriores e fornecer uma referência rápida para comparação entre os modelos avaliados.

## CNN Baseline

A CNN Baseline foi desenvolvida especificamente para este trabalho e treinada inteiramente a partir das imagens disponíveis no conjunto de dados. Sua utilização teve como finalidade estabelecer uma referência mínima de desempenho, permitindo quantificar os ganhos proporcionados por arquiteturas mais sofisticadas.

Como esperado, o modelo demonstrou capacidade de aprender padrões relevantes das imagens, porém apresentou desempenho inferior aos modelos pré-treinados e indícios de sobreajuste após algumas épocas de treinamento. Ainda assim, sua simplicidade arquitetural e baixo custo computacional tornam essa abordagem adequada para fins didáticos e para problemas com restrições severas de processamento.

---

## MobileNetV2

A MobileNetV2 é uma arquitetura baseada em Transfer Learning originalmente treinada sobre o conjunto de imagens ImageNet. Seu principal diferencial é a utilização de convoluções separáveis em profundidade (*Depthwise Separable Convolutions*), permitindo reduzir significativamente o número de parâmetros sem comprometer o desempenho.

Nos experimentos realizados, a MobileNetV2 apresentou excelente equilíbrio entre desempenho, estabilidade e custo computacional, superando amplamente a CNN Baseline. Também foram conduzidos experimentos adicionais envolvendo ajuste da taxa de aprendizado, Data Augmentation e Fine Tuning, os quais contribuíram para uma compreensão mais aprofundada do comportamento do modelo, embora não tenham produzido ganhos significativos em relação à configuração original.

---

## ResNet50

A ResNet50 representa uma das arquiteturas clássicas mais importantes da literatura de Visão Computacional. Seu principal diferencial é o uso de conexões residuais (*Residual Connections*), que permitem o treinamento eficiente de redes profundas ao minimizar problemas de desaparecimento do gradiente.

Durante os experimentos iniciais, a ResNet50 apresentou desempenho competitivo, superando a MobileNetV2 tanto em acurácia quanto em capacidade de generalização. Posteriormente, foram realizados experimentos exploratórios de Fine Tuning, confirmando que a arquitetura possuía elevada capacidade de adaptação ao domínio específico das imagens de embarcações.

---

## YOLO11n-cls

A arquitetura YOLO11n-cls representa a versão voltada para classificação de imagens da família YOLO11, desenvolvida pela Ultralytics. Embora a família YOLO seja tradicionalmente associada à detecção de objetos em tempo real, sua variante para classificação incorpora técnicas modernas de Deep Learning capazes de produzir excelente desempenho com reduzido custo computacional.

Nos experimentos realizados, o YOLO11n-cls apresentou o melhor desempenho global entre todas as arquiteturas avaliadas, alcançando aproximadamente **95,1% de acurácia no conjunto de validação** e **92,96% de acurácia no conjunto de teste**, além de **Precision Macro de 0,9394**, **Recall Macro de 0,9371** e **F1-score Macro de 0,9379**.

A análise da matriz de confusão mostrou que a maior parte das classificações incorretas concentrou-se entre as categorias **Cargo** e **Tankers**, embarcações que apresentam elevada similaridade visual. Ainda assim, o modelo manteve desempenho consistente para todas as categorias avaliadas, demonstrando excelente capacidade de generalização.
